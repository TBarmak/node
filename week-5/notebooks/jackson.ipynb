```python
import pandas as pd
import numpy as np
import plotly #logistic regression
import plotly.express as px
import warnings
warnings.filterwarnings("ignore")
```

# Reading in Data

Reading in a dataset from Kaggle that has data on diamond 1 and 2 ranked games.

```python
league = pd.read_csv('high_diamond_ranked_10min.csv')
league
```

Based on my experience with League of Legends, there are many factors I'd expect would be able to predict a win or loss, but I want to know the strongest determinants.

# Questions:
> Which side wins more often, Red or Blue?

> How often does the team with first dragon win?

> How often does the team with a gold lead at 10 min win?

> How often does the team with an experience lead at 10 min win?

> What about combining all three factors?

> What's more effective at getting a gold/experience lead? Minions or kills?





```python
blue_wins = league['blueWins'].sum() #num of blue wins
red_wins = league.shape[0] - blue_wins #num of red wins
blueRed = pd.DataFrame() #create a new df for bar graph
blueRed['Team'] = ['Blue', 'Red']
blueRed['Wins'] = [4930, 4949]
blueRed['Win Percentage'] = [49.9, 50.1] #set up columns, i got these number by viewing the variables above
blueRedBar = px.bar(blueRed, x='Team', y='Win Percentage', color='Team', range_y=(49, 51), title='Red Side v. Blue Side Win Percentage') #set up a bar graph with very small range to view the slight difference
blueRedBar
```

This very small difference is expected as the two sides should be equal in strength.

```python
bluedragwins = league[league.blueDragons == 1][league.blueWins == 1]
reddragwins = league[league.redDragons == 1][league.blueWins == 0]
bluedragloss = league[league.blueDragons == 1][league.blueWins == 0]
reddragloss = league[league.redDragons == 1][league.blueWins == 1] #set up tables with the desired traits (red had dragon and lost for this one)
```

```python
drag = pd.DataFrame() #create table for bar graph
drag['First Dragon?'] = ['Yes', 'No'] #whether or not the team had first dragon
drag['Wins'] = [bluedragwins.shape[0] + reddragwins.shape[0], bluedragloss.shape[0] + reddragloss.shape[0]] #the number of times that team won
drag['Win %'] = [drag.Wins[0] / (drag.Wins[0] + drag.Wins[1]) * 100, drag.Wins[1] / (drag.Wins[0] + drag.Wins[1]) * 100,] #the percentage of games the team won
dragBar = px.bar(drag, x='First Dragon?', y='Win %', color='First Dragon?', title='Win % With and Without First Dragon') # bar graph
dragBar
```

The dragon is the only major objective to come up within the first 10 minutes and gives the team a buff that will impact the rest of the game so it makes sense for this to be a strong difference maker.

```python
blueGoldWins = league[league.blueGoldDiff > 0][league.blueWins == 1]
redGoldWins = league[league.redGoldDiff > 0][league.blueWins == 0]
blueGoldLoss = league[league.blueGoldDiff > 0][league.blueWins == 0]
redGoldLoss = league[league.redGoldDiff > 0][league.blueWins == 1] #pretty identical logic to above but for a gold difference instead of dragon
```

```python
Gold = pd.DataFrame()
Gold['Gold Lead?'] = ['Yes', 'No']
Gold['Wins'] = [blueGoldWins.shape[0] + redGoldWins.shape[0], blueGoldLoss.shape[0] + redGoldLoss.shape[0]]
Gold['Win %'] = [Gold.Wins[0] / (Gold.Wins[0] + Gold.Wins[1]) * 100, Gold.Wins[1] / (Gold.Wins[0] + Gold.Wins[1]) * 100,]
GoldBar = px.bar(Gold, x='Gold Lead?', y='Win %', color='Gold Lead?', title='Win % With and Without a Gold Lead at 10 min')
GoldBar
```

Since gold is arguably the best measure of a team's strength this factor should be the strongest in determining a win or a loss.

```python
blueXPWins = league[league.blueExperienceDiff > 0][league.blueWins == 1]
redXPWins = league[league.redExperienceDiff > 0][league.blueWins == 0]
blueXPLoss = league[league.blueExperienceDiff > 0][league.blueWins == 0]
redXPLoss = league[league.redExperienceDiff > 0][league.blueWins == 1] #pretty identical logic to above but for a xp difference instead of gold
```

```python
xp = pd.DataFrame()
xp['XP Lead?'] = ['Yes', 'No']
xp['Wins'] = [blueXPWins.shape[0] + redXPWins.shape[0], blueXPLoss.shape[0] + redXPLoss.shape[0]]
xp['Win %'] = [xp.Wins[0] / (xp.Wins[0] + xp.Wins[1]) * 100, xp.Wins[1] / (xp.Wins[0] + xp.Wins[1]) * 100,]
xpBar = px.bar(xp, x='XP Lead?', y='Win %', color='XP Lead?', title='Win % With and Without an XP Lead at 10 min')
xpBar
```

Similarly to gold, xp is another measure of a team's strength and shows similar strength as a determinant for a win or a loss.

```python
blue3Wins = league[league.blueExperienceDiff > 0][league.blueGoldDiff > 0][league.blueDragons == 1][league.blueWins == 1]
red3Wins = league[league.redExperienceDiff > 0][league.redGoldDiff > 0][league.redDragons == 1][league.blueWins == 0]
blue3Loss = league[league.blueExperienceDiff > 0][league.blueGoldDiff > 0][league.blueDragons == 1][league.blueWins == 0]
red3Loss = league[league.redExperienceDiff > 0][league.redGoldDiff > 0][league.redDragons == 1][league.blueWins == 1] #pretty identical logic to above but for all three factors
```

```python
three = pd.DataFrame()
three['Dragon with an XP & Gold Lead?'] = ['Yes', 'No']
three['Wins'] = [blue3Wins.shape[0] + red3Wins.shape[0], blue3Loss.shape[0] + red3Loss.shape[0]]
three['Win %'] = [three.Wins[0] / (three.Wins[0] + three.Wins[1]) * 100, three.Wins[1] / (three.Wins[0] + three.Wins[1]) * 100,]
threeBar = px.bar(three, x='Dragon with an XP & Gold Lead?', y='Win %', color='Dragon with an XP & Gold Lead?', title='Win % With and Without Dragon, Gold, and XP Lead at 10 min')
threeBar
```

After looking at all three of these factors, it appears that the dragon is less important than gold and experience. But there is somewhat of a debate in the community about the importance of kills and minions for acheiving a gold lead. Similarly how does dragon impact getting the gold lead?

```python
px.scatter(league, y='blueGoldDiff', x='blueTotalMinionsKilled', color='blueDragons', opacity=.5,
           title = 'Minions Killed Against the Gold Difference Between Teams',
           labels={'blueTotalMinionsKilled':'# Minions Killed', 'blueGoldDiff':'Gold Advantage'})
```

```python
px.scatter(league, y='blueGoldDiff', x='blueKills', color='blueDragons', opacity=.5, 
           title = 'Enemies Killed Against the Gold Difference Between Teams',
           labels={'blueKills':'# Enemies Slain', 'blueGoldDiff':'Gold Advantage'})
```

```python
px.scatter(league, y='blueGoldDiff', x='blueDragons', title = 'Dragons Killed Against the Gold Difference Between Teams',
           labels={'blueDragons':'# Dragons Killed', 'blueGoldDiff':'Gold Advantage'})
```

Neither of these scatterplots are especially revealing when it comes to which has a greater impact, minions or kills, but it does seem to show that getting the first dragon doesn't influence the gold difference much which may indicate that its importance is overvalued.

To try and get a better idea about the relative importance of kills and minions lets look at some regression output.

```python
import statsmodels.api as sm

X = league["blueTotalMinionsKilled"]
y = league["blueGoldDiff"]

# Note the difference in argument order
model = sm.OLS(y, X).fit()
predictions = model.predict(X) # make the predictions by the model

# Print out the statistics
model.summary()
```

```python
X = league["blueKills"]
y = league["blueGoldDiff"]

# Note the difference in argument order
model = sm.OLS(y, X).fit()
predictions = model.predict(X) # make the predictions by the model

# Print out the statistics
model.summary()
```

We'd expect the kills coeffecient to be much higher because there are much fewer kills than minion kills, however, the test statistics (f and t) are both much higher for kills than minions. Both variables would be considered significant in achieving a gold lead but it seems that kills are actually more significant. 

Does it differ for experience lead?

```python
X = league["blueKills"]
y = league["blueExperienceDiff"]

# Note the difference in argument order
model = sm.OLS(y, X).fit()
predictions = model.predict(X) # make the predictions by the model

# Print out the statistics
model.summary()
```

The results for experience difference are the same as for the gold difference. I would guess that in games where a team had more minions killed, there would be less kills overall and the game would be more even hence the less significance on the variable for difference between two teams but what about just accumulating gold and experience?

```python
X = league["blueTotalMinionsKilled"]
y = league["blueTotalExperience"]

# Note the difference in argument order
model = sm.OLS(y, X).fit()
predictions = model.predict(X) # make the predictions by the model

# Print out the statistics
model.summary()
```

```python
X = league["blueKills"]
y = league["blueTotalExperience"]

# Note the difference in argument order
model = sm.OLS(y, X).fit()
predictions = model.predict(X) # make the predictions by the model

# Print out the statistics
model.summary()
```

As expected when just looking at the sheer amount of gold and experience, it appears the number of minions killed is a much stronger indicator than kills on enemy players. However, winning the game usually requires getting leads over the opponent not just matching their strength so in that sense, kills should be valued more as they're better for creating differences between the two teams. As stated before though, both are definitely significant in accumulating gold, experience, and leads over the opponent so neither's importance should be ignored.
