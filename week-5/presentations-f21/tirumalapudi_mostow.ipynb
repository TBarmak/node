```python
income = pd.read_csv("incomegroups.csv", sep='\n')
income.head(5)
income
```

```python
Income = income['Country Code,"Region","IncomeGroup","SpecialNotes","TableName",'].str.split('"',expand=True)
Income
```

```python
Income = Income.drop(columns=[0,1,2,4,5,6,8])
```

```python
Income = Income.rename(columns={3: 'IncomeLevel', 7:'Country Name'})
```

```python
table = pd.merge(finaltable, Income, how = 'inner', on = 'Country Name')
```

```python
table
```

```python
table['GDP growth (annual % for 2014)'] = table['GDP growth (annual % for 2014)'].astype(float)
table['Government expenditure on education, total (% of GDP)'] = table['Government expenditure on education, total (% of GDP)'].astype(float)
table['Literacy rate (% of females ages 15 and above)'] = table['Literacy rate (% of females ages 15 and above)'].astype(float)
```

```python
table
```

```python
#groupby income level
countries_byI = table.groupby(by='IncomeLevel')
medians = countries_byI['Literacy rate (% of females ages 15 and above)'].agg('mean', 'median')
```

```python
#px.bar(table, x=countries_byI, y=medians)
```

```python
px.histogram(table, x='Literacy rate (% of females ages 15 and above)', color='IncomeLevel', range_x=(0,100), range_y=(0,5) )
#key takeaway is the differences in shape and spread for the lower -> higher income groups, literacy rate undeniably increases with income but
#the distributions show that there is more to it than that (i.e. low income literacy cannot be predicted solely from income)
```

```python
expenditure_vs_lit = px.scatter(table, x='Government expenditure on education, total (% of GDP)', y='Literacy rate (% of females ages 15 and above)', color = 'IncomeLevel',range_y=(20,100), range_x=(2,6))
```

```python
gdpgrowth_vs_lit = px.scatter(table, x='GDP growth (annual % for 2014)', y='Literacy rate (% of females ages 15 and above)', color= 'IncomeLevel', range_y=(20,100), range_x=(-2,8), hover_name='Country Name',)
```

```python
gdpgrowth_vs_lit
#low income is the ONLY category that shows a significant relationship- therefore an improvement to the economic conditions
#of a country does not correlate to women's educational quality in the same way across income levels
```

```python
expenditure_vs_lit
#by contrast, all income levels show at least some sort of relationship between rel expenditure and women's literacy
#what does this mean? one possible explanation is that more of the available money for gov't spending as a result of gdp growth
#is used on education and other basic needs of a country in lower income countries
#we can test this hypothesis by plotting gdp growth against expenditure!
```

```python
gdp_vs_exp = px.scatter(table, x='GDP growth (annual % for 2014)', y='Government expenditure on education, total (% of GDP)', color= 'IncomeLevel', range_y=(0,8), range_x=(-2,8), hover_name='Country Name')
```

```python
gdp_vs_exp
#not only are lower income countries showing a clear relationship between the two variables, its the ONLY category that has any kind of
#discernable pattern!! This is important because the predictive ability of gdp growth -> expenditure completely dissolves even
#at the lower middle income level, which many people would mistakenly assume follows a similar pattern to low income countries
```

```python
#Why are these findings important? Discerning this key difference between low income and every other category is crucial to meaningful and
#effective activism toward improving women's education. Our approach should not be the same for countries with different characteristics
#for low income, the crucial piece is that more funding is available for education, whereas in upper categories it is the general income level
#and gdp growth that predict literacy best
```

```python
#In countries where the basic needs are met, literacy comes down to other factors that could be further analyzed- for example:
#degree of income inequality, social dynamics (role of women in society), and other cultural factors
#Our data essentially begins to uncover the nuances of women's education throughout varying conditions, and provides insight as to the best
#course of action if we are aiming to improve it
```
